# Binning

```python
import pandas as pd
file = pd.read_csv("binnig data.csv")
file
```

```python
bin = [0 , 20 , 35 , 50 , 70]    #yeh hum ny grouping ki ha jis ko dekhty howy hamary groups bany gay 0  - 20  first group , 20 - 35 2nd group m 35 - 50  third group , 50 - 70 4rth group 
group_name = ["Fail" , "Pass" , "Good" , "Excellent"]    #yeh wo names hain jo un group ki values k samny likhian jain gyy
file["Grades"] = pd.cut(file["Marks"] , bin , labels = group_name)   #es main hamny sub sy pehly 1 extra column bnaya jis ka name grade ha us k bad 
file
```
