```python
import pandas as pd 
df = pd.read_csv("profit data.csv")
df
```

```python
data = df.groupby("cities")
data
```

```python
for city_name , city_profit_information in data:
    print(city_name)   #yeh sirf city name print kry ga
    print(city_profit_information)     #yeh us city ki full information print kry ga
```

```python
print(data.get_group("Multan"))
data.get_group("Taxila")     #yeh kisi 1 city ka data find krny k liye use krty hain
```

```python
print(data.min())    #yeh har city main sy jo minimum value hogi print krwa dy ga
print(data.max())
print(data.mean())         #yeh interger value column main mean find krny k liye
data.median()            ##yeh interger value column main median find krny k liye
```

```python
data.describe()    #yeh puray data ko achi trah sy describe krta ha
```
